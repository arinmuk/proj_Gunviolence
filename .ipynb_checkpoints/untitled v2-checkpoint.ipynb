```python
import matplotlib.pyplot as plt
import pandas as pd
from functools import reduce
import numpy as np
import requests
import time
import datetime 
from datetime import date 
import pprint as pp
file="gun-violence-data_01-2013_03-2018.csv"
```

```python
#['incident_id', 'date', 'state', 'city_or_county','n_killed',
#       'n_injured',  'latitude',
#       'location_description', 'longitude','participant_age', 'participant_age_group', 'participant_gender']



fields=['incident_id',
 'date',
 'state',
 'city_or_county',
 'n_killed',
 'n_injured','latitude',
 'longitude', 'participant_age',
 'participant_age_group',
 'participant_gender',
 'participant_status',
 'participant_type']
df=pd.read_csv(file,usecols=fields)
df.head(12)
```

```python
df.count()
```

```python
#creating subset file for graps related to crime and region and victim counts. cleanup phase 1
inc_dt_region_df = df[['incident_id', 'date','state','city_or_county','n_killed','n_injured']].copy()
inc_dt_region_df.head()

inc_dt_region_df.to_csv("inc_dt_region_vict.csv",index=False) 
del inc_dt_region_df
```

```python
#creating csv for lat long plot . remove the null values for lat long. approx 8k rows were eliminated that had no lat long
#Cleanup Ph 2
inc_dt_reg_lat_lon_df = df[['incident_id', 'date','state','latitude','longitude']].copy()
inc_dt_reg_lat_lon_df =inc_dt_reg_lat_lon_df.dropna(how='any')
inc_dt_reg_lat_lon_df.to_csv("inc_dt_reg_lat_lon.csv",index=False) 
inc_dt_reg_lat_lon_df.count()
inc_dt_reg_lat_lon_df.head()

del inc_dt_reg_lat_lon_df
```

```python

#split and get the invidual participant ages as separate rows if there are multiple participants and their ages 
#in particanpnt age field
working_df = df[['incident_id','date', 'participant_age','state','city_or_county','n_killed','n_injured']].copy()
working_df=working_df.dropna(how='any')
list1_part_age=[]
list2_part_age=[]
master_age_ls=[]
incidentid=0
for index, row in working_df.iterrows():
    #incidentid = row['incident_id']
    part_age=row['participant_age']
    list1_part_age=part_age.split('||')
    for item in list1_part_age:
        list2_part_age=item.split('::')
        list2_part_age.append(row['incident_id'])
        list2_part_age.append(row['date'])
        list2_part_age.append(row['state'])         
        list2_part_age.append(row['city_or_county']) 
        list2_part_age.append(row['n_killed']) 
        list2_part_age.append(row['n_injured'])      
                 
                 
        master_age_ls.append(list2_part_age)
        #print(list2_part_age_grp)
age_df=pd.DataFrame(master_age_ls,columns=['part_age_cnt','age','incident_id','date','state','city_or_county','n_killed','n_injured'])    
#age_grp_df['incident_id']=age_grp_df['incident_id'].astype(str)
age_df.head(24)
#print(master_ls)
#**********************************************##########################
#participant_gender	participant_status	participant_type cleanup for these fields

part_gen_stat_type_df=  df[['incident_id','date','participant_gender','state','city_or_county','n_killed','n_injured']].copy()
part_gen_stat_type_df=part_gen_stat_type_df.dropna(how='any')
#### Gender Field work

#genlist1=[]
genlist1=[]
genlist2=[]
genmaster_ls=[]
incidentid=0
for index, row in part_gen_stat_type_df.iterrows():
    #incidentid = row['incident_id']
    part_gender=row['participant_gender']
    #part_status=row['participant_status']
    genlist1=part_gender.split('||')
    for item in genlist1:
         
        genlist2=item.split('::')
        
        genlist2.append(row['incident_id'])
        genlist2.append(row['date'])
        genlist2.append(row['state'])         
        genlist2.append(row['city_or_county']) 
        genlist2.append(row['n_killed']) 
        genlist2.append(row['n_injured'])  
        genmaster_ls.append(genlist2)

gender_df=pd.DataFrame(genmaster_ls,columns=['gender_cnt','gender','incident_id','date','state','city_or_county','n_killed','n_injured'])    
#age_grp_df['incident_id']=round(age_grp_df['incident_id']).astype(str)
gender_df.head(24)        
##### genderfield work complete######################
```

```python
gender_age_merged_df=age_df.merge(gender_df, left_on=['incident_id','part_age_cnt'], right_on=['incident_id','gender_cnt'], how='outer')
gender_age_merged_df.head(12)
```
