```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# File to Load (Remember to change these)
filetoread = "inc_dt_region_vict.csv"


# Read the City and Ride Data
region_vict_df = pd.read_csv(filetoread)
#Convert the data type of Date of Incident
region_vict_df['Date of Incident'] = pd.to_datetime(region_vict_df['date'], errors='coerce')
region_vict_df['Year'] = region_vict_df['Date of Incident'].dt.year
region_vict_df['Month'] = region_vict_df['Date of Incident'].dt.month
region_vict_df['Day'] = region_vict_df['Date of Incident'].dt.day
region_vict_df['casualty'] = region_vict_df['n_killed']+region_vict_df['n_injured']

city_violence_df=region_vict_df.groupby(["Year",'state',"city_or_county"]).agg({'incident_id':'count', 'casualty':'sum'})
#,as_index=False
city_violence_df.head()
#city_violence_df.to_csv("cityviotest.csv",index="False")
```

```python
# building least adn worst gun violence prone cities by each year


year_city_vio_df=city_violence_df.loc[2013,["incident_id","casualty"]]
year_city_vio_df["Year"]=2013
year_city_vio_df["incident_id"]=year_city_vio_df["incident_id"].astype(int)
year_city_vio_df["casualty"]=year_city_vio_df["casualty"].astype(int)
year_city_vio_df=year_city_vio_df.reset_index()
year_city_vio_df=year_city_vio_df.sort_values(by=["incident_id","casualty"], ascending=False)
year_city_vio_df=year_city_vio_df.reset_index()
worst2013_city=year_city_vio_df.iloc[[0,1,2,3,4]]
best2013_city=year_city_vio_df.iloc[[-1,-2,-3,-4,-5]]

year_city_vio_df=city_violence_df.loc[2014,["incident_id","casualty"]]
year_city_vio_df["Year"]=2014
year_city_vio_df["incident_id"]=year_city_vio_df["incident_id"].astype(int)
year_city_vio_df["casualty"]=year_city_vio_df["casualty"].astype(int)
year_city_vio_df=year_city_vio_df.reset_index()
year_city_vio_df=year_city_vio_df.sort_values(by=["incident_id","casualty"], ascending=False)
year_city_vio_df=year_city_vio_df.reset_index()
worst2014_city=year_city_vio_df.iloc[[0,1,2,3,4]]
best2014_city=year_city_vio_df.iloc[[-1,-2,-3,-4,-5]]


year_city_vio_df=city_violence_df.loc[2015,["incident_id","casualty"]]
year_city_vio_df["Year"]=2015
year_city_vio_df["incident_id"]=year_city_vio_df["incident_id"].astype(int)
year_city_vio_df["casualty"]=year_city_vio_df["casualty"].astype(int)
year_city_vio_df=year_city_vio_df.reset_index()
year_city_vio_df=year_city_vio_df.sort_values(by=["incident_id","casualty"], ascending=False)
year_city_vio_df=year_city_vio_df.reset_index()
worst2015_city=year_city_vio_df.iloc[[0,1,2,3,4]]
best2015_city=year_city_vio_df.iloc[[-1,-2,-3,-4,-5]]


year_city_vio_df=city_violence_df.loc[2016,["incident_id","casualty"]]
year_city_vio_df["Year"]=2016
year_city_vio_df["incident_id"]=year_city_vio_df["incident_id"].astype(int)
year_city_vio_df["casualty"]=year_city_vio_df["casualty"].astype(int)
year_city_vio_df=year_city_vio_df.reset_index()
year_city_vio_df=year_city_vio_df.sort_values(by=["incident_id","casualty"], ascending=False)
year_city_vio_df=year_city_vio_df.reset_index()
worst2016_city=year_city_vio_df.iloc[[0,1,2,3,4]]
best2016_city=year_city_vio_df.iloc[[-1,-2,-3,-4,-5]]

year_city_vio_df=city_violence_df.loc[2017,["incident_id","casualty"]]
year_city_vio_df["Year"]=2017
year_city_vio_df["incident_id"]=year_city_vio_df["incident_id"].astype(int)
year_city_vio_df["casualty"]=year_city_vio_df["casualty"].astype(int)
year_city_vio_df=year_city_vio_df.reset_index()
year_city_vio_df=year_city_vio_df.sort_values(by=["incident_id","casualty"], ascending=False)
year_city_vio_df=year_city_vio_df.reset_index()
worst2017_city=year_city_vio_df.iloc[[0,1,2,3,4]]
best2017_city=year_city_vio_df.iloc[[-1,-2,-3,-4,-5]]

year_city_vio_df=city_violence_df.loc[2018,["incident_id","casualty"]]
year_city_vio_df["Year"]=2018
year_city_vio_df["incident_id"]=year_city_vio_df["incident_id"].astype(int)
year_city_vio_df["casualty"]=year_city_vio_df["casualty"].astype(int)
year_city_vio_df=year_city_vio_df.reset_index()
year_city_vio_df=year_city_vio_df.sort_values(by=["incident_id","casualty"], ascending=False)
year_city_vio_df=year_city_vio_df.reset_index()
worst2018_city=year_city_vio_df.iloc[[0,1,2,3,4]]
best2018_city=year_city_vio_df.iloc[[-1,-2,-3,-4,-5]]



#worst2014_city.head()
#best2018_city.head()
#year_city_vio_df.head(10)
#year_city_vio_df.columns

worst_city_concat_df=pd.concat([worst2013_city,worst2014_city,worst2015_city,worst2016_city,worst2017_city,worst2018_city])
del worst_city_concat_df['index']
worst_city_concat_df.set_index("Year",inplace=True)
worst_city_concat_df.head(20)
```

```python

ax_df=worst_city_concat_df.pivot(index=worst_city_concat_df.index, columns='city_or_county').plot.bar(rot=1,subplots=False,Width=.15,edgecolor='black',position=0,legend='city_or_county',figsize=(12,9),align="center",sort_columns=True)
#ax_df.plot.bar(rot=0,subplots=False,legend='')
plt.title("Gun Violence-Top 5 Cities ")

plt.ylabel("Casualties(Killed/Injured)")
lgnd=plt.legend(loc='upper right',markerscale=.5,title="city_or_county")
#plt.figure(figsize=(10,10))

plt.savefig("5 worst cities-taran.png")
plt.show()
```
