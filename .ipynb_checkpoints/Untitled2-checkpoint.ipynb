```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# File to Load (Remember to change these)
gender_age_filetoread = "gender_age.csv"


# Read the City and Ride Data
agegroup_df = pd.read_csv(gender_age_filetoread)


agegroup_df.head()
```

```python
#Convert the data type of Date of Incident
agegroup_df['Date of Incident'] = pd.to_datetime(agegroup_df['date'], errors='coerce')
agegroup_df['Year'] = agegroup_df['Date of Incident'].dt.year
agegroup_df['Month'] = agegroup_df['Date of Incident'].dt.month
agegroup_df['Day'] = agegroup_df['Date of Incident'].dt.day
agegroup_df['casualty'] = agegroup_df['n_killed']+agegroup_df['n_injured']
```

```python
bins = [0, 9, 19, 29, 39, 49,59,69,100]



group_names = ["<10", "10-19", "20-29", "30-39", "40-49", "50-59", "60-69","70+"]
agegroup_df["age"] = pd.cut(agegroup_df["age"], bins, labels=group_names)
agegroup_df.head()
```

```python
agegroup_grp=agegroup_df.groupby(["Year","age"],as_index=False)["incident_id"].count()
agegroup_grp.head()
```

```python
ax_df=agegroup_grp.pivot_table(index='Year', columns='age',values=['incident_id'])
ax_df.head()
```

```python
ax_df.plot.bar(rot=1,subplots=False,Width=.25,edgecolor='black',stacked=True,colormap='Paired',position=0,legend='Age groups Involved',figsize=(12,9),sort_columns=True)
```
