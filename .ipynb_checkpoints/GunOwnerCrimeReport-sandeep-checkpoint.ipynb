```python
# import the libraries
%matplotlib inline
#import matplotlib as plt
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#import seaborn as sns
import time 

```

```python
# Initialize variables. Read data files
#file_to_parse_gunown = "../DataGun/GunOwnershipCSV.csv"
#file_to_parse_casualty = "../DataGun/DF_GunDataFull.csv"
#file_to_parse_population = "../DataGun/population.csv"

file_to_parse_gunown = "gunowners.csv"
file_to_parse_casualty = "inc_dt_region_vict.csv"
file_to_parse_population = "statepopulation.csv"


df_gun_ownership = pd.read_csv(file_to_parse_gunown)
df_gun_casualty  = pd.read_csv(file_to_parse_casualty)
df_population = pd.read_csv(file_to_parse_population)
df_gun_casualty.head()


```

```python
# renaming fields to match joins

df_gun_casualty["date"] = pd.to_datetime(df_gun_casualty["date"])
df_gun_casualty["date"]= df_gun_casualty["date"].dt.year
df_gun_ownership = df_gun_ownership.rename(columns={"State":"state"})
df_population = df_population.rename(columns={"States":"state"})

df_gun_casualty.head()

```

```python
#Aggreagtion of data frames  get incident counts , casulaty by state
df_gun_casualty_agg = df_gun_casualty.groupby(['state'],as_index= False).agg({'incident_id':'count','n_killed':'sum','n_injured':'sum'})
df_gun_casualty_agg["Total Casualty"] =df_gun_casualty_agg["n_killed"] + df_gun_casualty_agg["n_injured"]

df_gun_casualty_agg.head()
```

```python
df_population.head()
```

```python
# Data Merge

df_combined_all = pd.merge(df_gun_casualty_agg,df_population, how='inner', on='state')
df_combined_all = pd.merge(df_combined_all,df_gun_ownership, how='inner', on='state')
#df_combined_selectedColumns = df_combined_all[df_combined_all.columns['state','date','population','# of guns registered','Total Casualty']]
df_combined_selectedColumns =df_combined_all[['state','Population','# of guns per capita','# of guns registered','Total Casualty']]
df_population_only =df_combined_all[['Population','state']]
#df_population_only
df_combined_selectedColumns["Population"]=df_combined_selectedColumns["Population"].str.replace(',', '')
df_combined_selectedColumns["Population"]=pd.to_numeric(df_combined_selectedColumns["Population"])
df_combined_selectedColumns["Total Calculated Registered Gun"] =df_combined_selectedColumns["# of guns per capita"] * df_combined_selectedColumns["Population"]
df_combined_selectedColumns["Total Calculated Registered Gun"] =(df_combined_selectedColumns["Total Calculated Registered Gun"] /1000).round(0)
df_combined_selectedColumns.head(52)
```

```python
#plotting charts

fig, ax = plt.subplots()

x_axis = np.arange(len(df_combined_selectedColumns))
tick_locations = [value+0.4 for value in x_axis]
#plt.figure(figsize=(40,6))
ax = plt.gcf().set_size_inches((26,11))
plt.xticks(tick_locations, df_combined_selectedColumns["state"], rotation="vertical",fontsize=24)
plt.yticks(fontsize=22)

#plt.stackplot(df_combined_selectedColumns['state'], [df_combined_selectedColumns['Total Casualty'],df_combined_selectedColumns['Total Calculated Registered Gun'],df_combined_selectedColumns['Population']],colors=["orange", "lightgreen", "lightblue"])
plt.stackplot(df_combined_selectedColumns['state'],df_combined_selectedColumns['Population'],colors=["lightblue"],edgecolors="black")
plt.stackplot(df_combined_selectedColumns['state'],df_combined_selectedColumns['Total Calculated Registered Gun']*20,colors=["orange"],edgecolors="black")
plt.stackplot(df_combined_selectedColumns['state'],df_combined_selectedColumns['Total Casualty']*200,colors=["pink"],edgecolors="black")
plt.xlabel('States', fontsize=22)
plt.ylabel('Count in Millions', fontsize=22)
#ax = plt.gcf().set_size_inches((10,10))
plt.title('State population vs Gun Ownership to Casualties ( Injured/Killed)',fontsize=26)

plt.grid()
plt.legend(['Population','Gun Ownership','Crimes'],prop={'size': 19})
plt.tight_layout()
plt.savefig("state population_reg-guns-casualty.png")
plt.show()                       
#loc=2, prop={'size': 6}
```
